### Expression:

(2 + 1) * 3

_*This is an example - you do not need to complete anything.*_

```python
from IPython.display import Image
import graphviz

# Create a new directed graph
dot = graphviz.Digraph(comment='Abstract Syntax Tree Example')
dot.attr(rankdir='TB')  # Top to bottom direction

# Create nodes for the expression.The order in which the nodes are created
# can affect whether subtrees appear on the left or right.
dot.node('*', '*')
dot.node('+', '+')
dot.node('2', '2')
dot.node('1', '1')
dot.node('3', '3')

# Create edges to connect nodes
dot.edge('*', '+')  # Edge: * -> +
dot.edge('+', '2')  # Edge: + -> 2
dot.edge('+', '1')  # Edge: + -> 1
dot.edge('*', '3')  # Edge: * -> 3

# Save and render the graph
dot.render('ast_example', view=False, format='png')
display(Image('ast_example.png'))
```
